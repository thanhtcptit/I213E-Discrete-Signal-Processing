```python
import numpy as np
import scipy    
import matplotlib.pyplot as plt
```

```python
K = 500
G_r, G_i = [], []
for k in range(K):
    G = 1 / (1 - 0.1 * complex(np.cos(2 * np.pi * k / K), np.sin(2 * np.pi * k / K)))
    G_r.append(G.real)
    G_i.append(G.imag)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(G_r)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(G_i)
plt.show()
```

```python
a = [1, -0.1]
b = [1]
k = list(range(120))

x = [np.cos(0.05 * np.pi * i) for i in k]
y = [1.109 * np.cos(0.05 * np.pi * i - 0.0173) for i in k]
y2 = scipy.signal.lfilter(b, a, x)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(y)
plt.show()
```
