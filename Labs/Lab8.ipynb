```python
import numpy as np
import scipy    
import matplotlib.pyplot as plt
```

```python
dt = 5e-5
t = np.arange(-0.005, 0.005, dt)[np.newaxis, :]
xa = np.exp(-1000 * np.abs(t))
```

```python
Fs = 2000
Wmax = 2 * np.pi * Fs
L = 500
i = np.arange(0, L)[np.newaxis, :]
W = i * Wmax / L
Xa = np.dot(xa, np.cos(np.dot(t.T, W))) * dt

W2 = np.hstack([-np.flip(W), W[:, 2:]])
Xa2 = np.hstack([np.flip(Xa), Xa[:, 2:]])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(t * 1000, xa, '-b.', linewidth=1, markersize=2)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(W2 / (2 * np.pi * 1000), Xa2 * 1000, '-b.', linewidth=1, markersize=2)
plt.show()
```
